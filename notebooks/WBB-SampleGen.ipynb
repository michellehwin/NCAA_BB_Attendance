```python
import json
import pandas as pd
```

```python
# create df from team ids
team_id_file = open('../espn_team_ids.json')
team_id_json = json.load(team_id_file)
team_id_df = pd.DataFrame.from_dict(team_id_json, orient='index')
team_id_df.reset_index(inplace=True)
team_id_df.columns=['Team_Name','ID']
team_id_df.set_index('ID',drop=True,inplace=True)
team_id_df
```

```python
team_sample = team_id_df.sample(n = 20)
team_sample
```

```python
from datetime import date
team_sample.to_csv(str(date.today())+'sample.csv')
```
