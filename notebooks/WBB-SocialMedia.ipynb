```python
import pandas as pd
from googlesearch import search
```

```python
# load df from csv
sample_df = pd.read_csv("2023-02-15sample.csv")
sample_df
```

```python
twitter_dict = []
instagram_dict = []
sample_df.reset_index()
for index in sample_df.index:
    twitter = "ERROR"
    instagram = "ERROR"
    query = f"{sample_df['Team_Name'][index]} wbb"
    twitter_query = search(query + " twitter", num=3)
    for result in twitter_query:
        if "twitter.com" in result:
            twitter = result[:result.find('?')]
            break
    ig_query = search(query + " instagram", num=3)
    for result in ig_query:
        if "instagram.com" in result:
            result_clean_index = result.find('?')
            if (result_clean_index > 0):
                instagram = result[:result.find('?')]
            else:
                instagram = result
            break
    twitter_dict.append(twitter)
    instagram_dict.append(instagram)
    print(twitter)
    print(instagram)
```

```python
socials_df = pd.DataFrame(
    {'Twitter': twitter_dict, 'Instagram': instagram_dict})
sample_df = sample_df.join(
    socials_df)
sample_df
```
